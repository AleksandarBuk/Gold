```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the monthly Gold prices dataset
monthly_data = pd.read_csv('monthly_csv.csv') 

monthly_data['Date'] = pd.to_datetime(monthly_data['Date'])

plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Date'], monthly_data['Price'], marker='o', linestyle='--')
plt.title('Monthly Gold Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

## 1. Trend Analysis

- Gold Prices Line: The blue line in the plot represents the actual monthly Gold prices. Each data point on the line corresponds to the Gold price for a specific month.

- 50-Month Moving Average (MA): The orange line on the plot represents the 50-month moving average of Gold prices. The moving average smooths out short-term fluctuations and highlights the long-term trend. When the Gold price line is above the 50-month MA, it suggests an upward trend, and when it's below, it indicates a downward trend.

```python
import pandas as pd
import matplotlib.pyplot as plt

monthly_data = pd.read_csv('monthly_csv.csv') 
monthly_data['Date'] = pd.to_datetime(monthly_data['Date'])

# Calculate a 50-month moving average
monthly_data['50_MA'] = monthly_data['Price'].rolling(window=50).mean()

# Create a line chart to visualize Gold prices and the 50-month moving average
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Date'], monthly_data['Price'], label='Gold Prices', alpha=0.7)
plt.plot(monthly_data['Date'], monthly_data['50_MA'], label='50-Month MA', color='orange')
plt.title('Monthly Gold Prices and 50-Month Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Volatility Analysis

- Volatility analysis measures the degree of variation or fluctuations in the price of an asset over time. In financial markets, higher volatility often implies higher risk and potential for larger price swings.


- Rolling 12-Month Volatility: The red line in the plot represents the rolling 12-month volatility of Gold prices. It shows how much the monthly returns of Gold prices have been fluctuating on a year-over-year basis. Higher peaks indicate periods of increased price volatility, while lower valleys represent relatively stable periods.

```python
# Calculate monthly returns
monthly_data['Returns'] = monthly_data['Price'].pct_change()

# Calculate rolling 12-month volatility
monthly_data['Rolling_Volatility'] = monthly_data['Returns'].rolling(window=12).std()

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Subplot 1: Rolling Volatility
ax1.plot(monthly_data['Date'], monthly_data['Rolling_Volatility'], label='Rolling Volatility (12-Month)', color='red')
ax1.set_title('Rolling 12-Month Volatility of Gold Prices')
ax1.set_ylabel('Volatility')
ax1.grid(True)
ax1.legend()

# Subplot 2: Monthly Returns
ax2.plot(monthly_data['Date'], monthly_data['Returns'], label='Monthly Returns', color='blue')
ax2.set_title('Monthly Returns of Gold Prices')
ax2.set_xlabel('Date')
ax2.set_ylabel('Returns')
ax2.grid(True)
ax2.legend()

# Adjust spacing
plt.tight_layout()
plt.show()
plt.savefig('Volatility Analysis')
```

## 3. Return Analysis

- Mean Monthly Return: This numeric value represents the average monthly return of Gold prices over the analyzed period. It provides insight into the typical monthly performance of Gold.

- Median Monthly Return: This numeric value represents the middle value of all monthly returns. It's a measure of central tendency and provides an alternative view of the typical monthly performance of Gold.

```python
# Calculate monthly cumulative returns
monthly_data['Cumulative_Returns'] = (1 + monthly_data['Returns']).cumprod() - 1

# Calculate mean and median returns
mean_return = monthly_data['Returns'].mean()
median_return = monthly_data['Returns'].median()

# Visualize cumulative returns
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Date'], monthly_data['Cumulative_Returns'], label='Cumulative Returns', color='green')
plt.title('Cumulative Returns of Gold Prices')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.grid(True)
plt.show()

print(f"Mean Monthly Return: {mean_return:.4f}")
print(f"Median Monthly Return: {median_return:.4f}")
```
